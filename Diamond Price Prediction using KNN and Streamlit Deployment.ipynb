### Define the input features (X) and the target variable (price).

```python
x = df.drop("price", axis =1)
y = df["price"]
```

### Identify Feature Types

```python
categorical_features = x.select_dtypes(include=['object']).columns
numerical_features = x.select_dtypes(include=['int64', 'float64']).columns

print("Categorical:", categorical_features)
print("Numerical:", numerical_features)
```

### Split the dataset into training and testing sets using a 75:25 ratio.

```python
# Preprocessing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.25, random_state=42)
```

```python
# Columns Transformer
from sklearn.compose import ColumnTransformer
preprocessor = ColumnTransformer(transformers=[("num", StandardScaler(), numerical_features),("cat", OneHotEncoder(handle_unknown="ignore"), categorical_features)])
```

### Build Full Pipeline (Preprocessing + Model)

```python
from sklearn.pipeline import Pipeline
pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                          ('model',KNeighborsRegressor(n_neighbors=5))])
```

```python
pipeline
```

### Train a KNN Regressor model using Scikit-learn

```python
# Model
from sklearn.neighbors import KNeighborsRegressor
pipeline.fit(x_train, y_train)
```

### Predictions

```python
y_pred = pipeline.predict(x_test)
y_pred
```

### Evaluate the model using the following metrics:
- Mean Absolute Error (MAE)
- Root Mean Squared Error (RMSE)
- R² Score


```python
### Mean Absolute Error
mae = mean_absolute_error(y_test, y_pred)
mae
```

```python
### Root Mean Squared Error
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
rmse
```

```python
### R² Score
r2 = r2_score(y_test, y_pred)
r2
```

### Save the trained pipeline using Pickle for future use.


```python
with open("diamond_price_model.pkl", "wb") as f:
    pickle.dump(pipeline, f)
```
